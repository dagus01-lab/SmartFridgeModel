# Libraries import

```python
from detector.data.dataset import CustomDataset
from detector.data.augmentations import RandomBlackSquares, RandomNoise, RandomHorizontalFlip, RandomVerticalFlip, RandomHorizontalRotation, RandomRotation 
from torchvision.transforms import v2
from detector.model.model import MyDetectionModel
import torch
import numpy as np
```

# Dataset initialization

```python
img_width = 640
img_height = 640
transformations = v2.Compose([
        v2.Resize((img_width, img_height)),
        RandomHorizontalFlip(),
        RandomVerticalFlip(),
        RandomHorizontalRotation(),
        RandomRotation(),
        RandomBlackSquares(),
        RandomNoise(),
])
train_dataset = CustomDataset('detector/data/train', transform=transformations)
test_dataset = CustomDataset('detector/data/test', transform=transformations)
valid_dataset = CustomDataset('detector/data/valid', transform=transformations)
```

# Transformations on dataset images

# Model initialization

```python
from detector.model.model import MyDetectionModelWrapper
batch_size = 32
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
model = MyDetectionModelWrapper(model_path='detector/model/yolov8n.pt', label_dict_path='detector/data/data.yaml', device=device)
epochs = 200
from torch.utils.data import DataLoader
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
```

# Model training

```python
history = model.train_on_dataset(train_dataset, valid_dataset, batch_size, device, epochs)
```
